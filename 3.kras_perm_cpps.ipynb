## Plot filter counts

```python
methods = ['Perm', 'KRAS', 'Perm+KRAS', 'Step-by-step',]
pass_perm_filter = [287304, 9, 185, 279264]
pass_kras_filter = [75, 294152, 291333, 18680]
pass_both_filter = [73, 3, 20, 17273]
```

```python
from utils import plot_style_utils

plot_style_utils.set_pub_plot_context(context="talk")
fig, ax = plot_style_utils.simple_ax()
plot_style_utils.prettify_ax(ax)

ax.bar(methods, pass_perm_filter, color=plot_style_utils.CAT_PALETTE[0])
ax.set_yscale('log')
ax.set_ylabel('Number of molecules')
ax.set_title('Pass permeability filter')

ax.set_ylim(1, 1e6)
ax.set_xticklabels(methods, rotation=30, ha='right')

plot_style_utils.save_for_pub(fig=fig, path='figures/kras_perm/pass_perm_filter')
```
